# Estructuras de datos

Las estructuras de datos son formas de organizar información para trabajar con ellas de manera eficiente.

- Listas -> ordenadas, modificadas, permiten duplicados
- Tuplas -> ordenadas, inmutables, permiten duplicados
- Sets -> no son ordenados, sin duplicados
- Diccionarios -> pares claves:valor

Nota:
- Pilas (FILO -> first in, last out)
- Colas (FIFO -> first in, first out)
- Listas enlazadas (o doblemente enlazadas)

## Listas
Las listas son coelecciones ordenadas que podemos modificar y almacenan elementos de distintos tipos de datos

```python
lista_de_compras = ["manzanas", "pan", "leche", "huevos", "mantequilla", "queso", "jamón"]
print(lista_de_compras)
```

```python
print(lista_de_compras[0])
```

```python
# Agregar un elemento a la lista .append()
lista_de_compras.append("sopa")
lista_de_compras
```

```python
# Agregar un elemento en una posición específica
lista_de_compras.insert(2, "plátano")
lista_de_compras
```

```python
print(lista_de_compras)
```

```python
# Para eliminar un elemento de una lista existe el remove y el pop
lista_de_compras.remove("plátano")
print(lista_de_compras)
```

```python
lista_de_compras.pop() # Sin nada dentro del paréntesis borra el último elemento
lista_de_compras.pop(4) #o puedes agarrar la posición que quieras pero iniciando desde el 0
# Nota: te retorna el número eliminado
```

```python
# Podemos recorrer el arreglo por medio de un for
for elementos in lista_de_compras:
  print(elementos)
```

```python
# len() saca el tamaño de la lista
print(len(lista_de_compras))
for i in range(len(lista_de_compras)):
  print(f"{i + 1} - {lista_de_compras [i]}")
```

```python
promedios = [55.6,97.5,23.6,97.5,72.4,100]
print(promedios)
promedios.sort() # sort() sirve para ordenar la lista de manera ascendente
print(promedios)
# ordenar una lista de manera descendente
promedios.sort(reverse = True)
print(promedios)
```

```python
# Podemos ingrear al último elemento de una lista
print(promedios[-1])
```

```python
# Encontrar el mayor y el menor de una lista
print(max(promedios))
print(min(promedios))
```

```python
# Declaración y uso de matrices
calificaciones = [[100,88,93], [99,56,72]]
print(calificaciones)
i = 1
for cali_i in calificaciones:
  print(f"\nAlumno {i}")
  j = 1
  for cali_j in cali_i:
    print(f" Cali {j}:{cali_j}")
    j += 1
  j = 0
  i+= 1
```

```python
# Unión
matutino = {20322, 22183, 123123, 53234}
vespertino = {20322, 111111, 2222222, 3333333}
#Union con .union
total_matrículas = matutino.union(vespertino)
print(total_matrículas)
#Union con |
total_matrículas = matutino | vespertino
print(total_matrículas)
```

```python
#Intersección
ca = {1234,121,646,12414,73434,124124, 675234}
al = {1234, 111, 222, 333, 444}
#intersección con .intersection
ca_al = ca.intersection(al)
print(ca_al)
#Intersección con &
ca_al = ca & al
print(ca_al)
```

```python
#Diferencia#Usando .difference
solo_cálculo =ca.difference(al)
print(solo_cálculo)
#Usando -
solo_cálculo = ca - al
print(solo_cálculo)
```

```python
nombre = {"Ángel", "Eliud", "Diana", "Juan", "Juan"}
# remove, pop
nombre.remove("Ángel") # Puedes eliminar el que tu quieras
print(nombre)
nombre.pop() #Elimina un elemento de manera aleatoria
print(nombre)
```

## Diccionarios
Los diccionarios son colecciones no ordenadas de pares clave:valor

```python
alumno = {
    # clave     valor
    "nombre": "Eliud Sánchez",
    "edad":21,
    "carrera": "LCC",
    "semestre": 7,
    "promedio":88,
    "calificaciones": [100,94, 94, 97, 100]
}
print(type(alumno))
alumno
```

```python
print("El alumno ", alumno["nombre"], "esta estudiando ", alumno["carrera"])
```

```python
# Agregar al diccionario
alumno["área preferida"] = "Ciencia de Datos"
alumno
```

```python
for valor in alumno:
  print(alumno[valor])
```

```python
for aspecto, valor in alumno.items(): #Devuelve una vista de todos los pares clave:valor del diccionario
  print(aspecto, ":", valor)
```

```python
#El método .clear() también usar aquí.
#Tenemos el .get que nos ayuda a obtener el valor (value) de cierta clave (key)
print(alumno.get("nombre"))
```

```python
alumno_1 = alumno.items()
print(alumno_1)
print(type(alumno_1))
alumno_lista = list(alumno_1)
print(alumno_lista)
print(type(alumno_lista))
```

```python
# El método .values() sirve para obtener los values del diccionario, también puede usar
alumno_v = alumno.values()
print(type(alumno_v))
alumno_v
```

```python
if "carrera" in alumno:
  print("Existe carrera")
else:
  print("No existe carrera")
if("dinero") in alumno:
  print("Alumno tiene dinero")
else:
  print("Alumno no tiene dinero")
```

```python
# .copy() puedes copiar todo el diccionario en una nueva variable
# Update -> sirve para cambiar los vales (.update())
alumno.update({"semestre":6})
```
