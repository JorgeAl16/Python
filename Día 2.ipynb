## Elif
Hagamos un programa en el que preguntamos la calificación del 0 al 100 y queremos saber su equivalencia en el sistema de evaluación estadounidense. Para simplicidad, basémonos en esto:
- A+ = 100
- A = 90 - 99
- B = 80 - 89
- C = 70 - 79
- D = 60 - 69
- F =< 60

```python
cal = 20
if cal >= 0 and cal <= 100:
  if cal == 100:
    print("A+")
  elif cal >= 90:
    print("A")
  elif cal >= 80:
    print("B")
  elif cal>= 70:
    print("C")
  elif cal >= 60:
    print("D")
  else:
    print("F")
else:
  print("Error, la calificación debe estar entre 0 y 100")
```

```python
# El módulo (%)
# El módulo es un operador matemático que te devuelve el residuo de una división
# 10 % 3 da 1, porque 10 dividido entre 3 cabe 3 veces y sobra 1
# 7 % 2 da 1, porque el 7 no es divisible de forma exacta entre 2
num = 9
if num % 3 == 0 or num % 5 == 0:
  print(f"El número {num} es múltiplo de 3 o de 5")

#Saber si el número es par o impar usando not
num = 6
if not num % 2 == 0:
  print(f"El número {num} es impar")
```

## Operadores lógicos con condicionales

```python
# and, or, not
pan = True
queso = True
mayonesa = True
jamón = False
print(pan and queso and mayonesa and jamón)
if not (pan and queso and mayonesa and jamón):
  print("No te puedes hacer el sandwich")
```

## Ciclo FOR
Los llamados ciclos for, nos ayuda a realizar acciones durante una cantidad determinada de veces

```python
# Imprimir los números del 0 al 4
for i in range(5): # toma 5 como el número final pero inicia siempre en 0
  print(i)
```

```python
for num in range(2,11):
  print(num)
```

```python
cadena = "Python"
for letra_por_letra in cadena:
  print(letra_por_letra)
```

```python
# También existe el range (inicio, fin, paso)
for i in range(2,21,5):
  print(i)
```

##While

```python
suma = 0
números = int(input("Ingresa el número para sumar o 0 para salir: "))
while números != 0:
  suma += números
  números = int(input("Ingresa el número para sumar o 0 para salir: "))
print(f"Tu suma es: {suma}")
```

```python
# Usar el while como for
número = 0
while número <= 12:
  print(número)
  número += 1
```
