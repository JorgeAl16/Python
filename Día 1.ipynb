Operaciones matemáticas

```python
# 4to Nivel de importancia
# Suma Ordinaria
print(3 + 4) # un 7
variable1,variable2 = 20,40 # 60
print(variable1 + variable2)

# Resta ordinaria
print(2 - 5) # -3

# Declaración de varibales
v1 = 10.4
v2 = 15
resultado = v2 - v1
print(resultado) #4.6
```

```python
# 3er Nivel de importancia
# Multiplicación Ordinaria
print("Multiplicación Ordinaria")
M1, M2 = 2,3.14
print(M1 * M2) # 6.28
# División Ordinaria
print("-------------")
print("División Ordinaria")
M1,M2,M3,M4,M5,M6 = 2,5,6,5,4,5
print(M1 / M2)
print(M3 / M4)
print(M5 / M6)
# División de piso
print("-------------")
print("División de piso")
print(M1// M2)
print(M3 // M4)
print(M5 // M6)

print(6/2 * 3)
```

```python
# Potencia
print("Potencias")
print(2 ** 3) # 8
P1,P2 = 7,2
print(P1 ** P2) # 49
```

```python
# Área de un círculo
pi = 3.14
r = 4
A = pi * (r ** 2)
```
