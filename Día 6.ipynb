```python
#Calculadora
class Calculadora:
  def __init__(self,num1,num2):
    self.num1 = num1
    self.num2 = num2
  def sumar(self):
    print(self.num1 + self.num2)
  def resta(self):
    print(self.num1 - self.num2)
  def multiplicar(self):
    print(self.num1 * self.num2)
  def dividir(self):
    if self.num2 == 0:
      print("Número no divisible")
    else:
      print(self.num1 / self.num2)
Calculadora = Calculadora(7,0)
Calculadora.dividir()
Calculadora.sumar()
Calculadora.multiplicar()
```

6. Encapsulamiento y propiedades

```python
class Cuenta:
  def __init__(self,saldo):
    self.__saldo = saldo #Privada
    #Sirve para proteger los datos internos de un objeto
cuenta = Cuenta(1000)
cuenta.Cuenta = 2000 #sí puedes editarlo
#print(cuenta.__saldo) #Error no se puede acceder directamente
```

```python
#Getters y setters
class Cuenta:
  def __init__(self,saldo):
    def .__saldo = saldo
    @property
    def saldo(self): #Getter
      return self.__saldo
    @saldo.setter #setter se ejecuta cuando hacemos c.saldo = valor
    def saldo(self, valor): #Permite leer el valor de __saldo usando el método (c.saldo)
        self.__saldo = valor
c = Cuenta(300)
print(c.saldo)
c.saldo = 4000
```

## 7. Herencias y abstracción

```python
#Herencia
class vehiculo:
  def __init__(self, marca, modelo):
    self.marca = marca
    self.modelo = modelo
  def descripción(self):
    return f"{self.marca}:{self.modelo}"
  def mover(self):
    return "El vehículo se está moviendo..."
#--------------------------------------------------
class coche(vehiculo): #subclase 1 (clase hija)
  def mover(self):
    return "El coche está conduciento"
#--------------------------------------------------
class Moto(vehiculo):
  def mover(self):
    return "La moto avanza sobre dos ruedas"
coche2 = vehiculo("BYD", "Dolphin")
coche = coche("Toyota", "Corolla")
moto = Moto("Yamaha", "R6")
print(coche.mover())
print(moto.mover())
print(coche2.mover())
```

##Composición HAS-A
por qué se llama composición?
porque estás componiendo (armando) un objeto grande usando otros objetos. Es un principio de POO donde:
- Una clase no hereda de otra, sino que contiene un objeto de esa otra clase

```python
class Banco:
  def __init__(self,dinero):
    self.dinero = dinero
  def ver_total(self):
    print(f"Ttotal en la cuenta: ${self.dinero}")
class Persona:
  def __init__(self, cuenta: Banco): # Indica que este parámetro debería ser una instancia en la clase banco
    self.cuenta = cuenta #Persona tiene un banco, no hereda de él
```
