# Existe un solo tema que TODO programador DEBE de saber

JSON es un formato para guardad y transmitir datos. Es muy usado en APIs, archivos de configuración y almacenamiento de información.
(Se parece a un diccionario de Python, pero es solo texto)

¿Para qué sirve?
- Sirve para convertir objetos de python
-> texto JSON y también texto JSON -> objetos de python

Esto se llama:
- Serializar: convertir un objeto de phyton en texto JSON
- Deserializar: convertir un texto JSON a un objeto usable de python

```python
# Funciones principales:
# 1) json.dums(obj): Convierte un diccionario u objeto de python en un string JSON
import json
obj = {"nombre": "Ana"
, "edad": 25}
print(type(obj))
s = json.dumps(obj, ensure_ascii = False)
print(s)
print(type(s))
```

```python
# json.loads(texto_json): Convierte un string JSON en un diccionario python
restaurar = json.loads(s)
print(restaurar)
print(type(restaurar))
# Al venir de una página/persona/archivo desconocid@ debes tener cuidado

'''
Pueden contener:
Datos malinformados
Información inesperada
Campos faltantes
Datos demasiado grandes
Variables incorrectas que rompan tu programa
'''
```

## Numpy
Provee el tipo ndarray, operaciones vectorizadas, brodcasting, utilidades de álgebra lineal y funciones matemáticas en arrays. Es la base sobre la que se apoyan muchas liberías cintíficas en python.

### Nota importante: el arreglo de Numpy no es igual a la lista de python, este NO puede tener diferentes tipos de datos en el mismo

```python
# pip install numpy
import numpy as np
A = np.array([[1,2,3], [4,5,6]]) #array (2,3)
V =np.array([10,20,30]) #array (3)
print(A + V)
print("----------------")
prueba = [1,2,3]
prueba2 = [1,2,3]
print(prueba + prueba2)
```

```python
#Más funciones de Numpy
z = np.zeros((3,3)) #array (3,3)
print(z)
print("----------------")
o = np.ones((2,4))
print(o)
print("-----------------")
r = np.random.rand(3,3)
print(r)
```

```python
#Rangos numéricos
nums = np.arange(0,10,2) #saca los valores entre 0 y 10 de dos en dos
print(nums)
x = np.linspace(0,1,5) #saca cinco valores espaciados de manera "equitativa" entre 0 y uno
print(x)
```

```python
a = np.array([1,2,3,4,5])
b = np.array([6,7,8,9,10])
print(a + b)
print (a - b)
print (a * b)
print (a / b)
print (a ** b)
```

```python
print(a.dot(b)) #Producto entre dos matrices
c = np.array([[1,2,3],[4,5,6]])
d = np.array([[10,20], [30,40], [50,60]])
#1*10 + 2*30 + 3*50
#
print(c.dot(d))
```

```python
data = np.array([1,2,3,4,5])
print(data.min()) #mínimo
print(data.max()) #máximo
print(data.suma)
print(data.mean()) #promedio
print(data.s)
```
