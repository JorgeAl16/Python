# Funciones
En este curso hemos usado algunas funciones integradas como python, pero nosotros también podemos definir funciones, siendo estas un bloque de código reutilizable que podemos mandar a llamar varias veces al código

```python
def mensaje():
  print("Hola!!")

mensaje()
```

```python
# Función con retorno
def mensaje():
  return "Hola!!"
print(mensaje())
va = mensaje()
print(va)
```

```python
def saludo(nombre):
  print(f"Hola", nombre, "!!")
tuNombre = input("Ingresa tu nombre: ")
saludo(tuNombre)
# Nota: El nombre de la variable del parámetro puede o no ser el mismo fuera de la ecuación
```

```python
def suma(número_1, número_2):
  return número_1 + número_2

suma = suma(1,2)
print(suma)
```

```python
# Descuento
def descuento(precio, desc = 0.15):
  precio = precio * (1 - desc)
  print("El nuevo precio es: ", precio)
descuento(750)
descuento (800, 0.25)
```

## Funciones que llaman a otras funciones

```python
def saludar(nombre):
  return(f"Hola, {nombre}!")

def despedirse(nombre):
  return(f"Adiós, {nombre}!")

def mensaje_completo(nombre):
  saludo = saludar(nombre)
  despedida = despedirse(nombre)
  return(saludo + "\n" + despedida)

print(mensaje_completo("Ana"))
```

```python
# Ahora supongamos que tenemos una heladería

def ask_flavor():
    flavors = ["vainilla", "chocolate", "strawberry"]
    print("Available flavors:", ", ".join(flavors))

    flavor = input("Choose a flavor: ").strip().lower()

    while flavor not in flavors:
        flavor = input("Choose a valid flavor: ").strip().lower()

    return flavor


def ask_quantity():
    # Pide la cantidad y valida que sea un entero positivo
    while True:
        try:
            quantity = int(input("How many cream balls?: "))
            if quantity > 0:
                return quantity
            else:
                print("Please enter a number greater than 0.")
        except ValueError:
            print("Please enter a valid integer.")


# Función madre (llama a las otras funciones)
def ask_cream():
    flavor = ask_flavor()
    quantity = ask_quantity()
    print(f"Order: {quantity} ball(s) of {flavor}")
    return quantity, flavor


# Programa principal
total_orders = int(input("How many customers will you serve?: "))

for i in range(total_orders):
    print(f"\n--- Customer {i+1} ---")
    ask_cream()
```

## Funciones Lambda
Una función lambda es una función acortada para escribir una función de esta forma:
nombreFunción = lambda parámetros: acción

La diferencia entre las funciones lambda y las funciones normales es que suelen ser un poco más limitadas

```python
sumar = lambda x,y : x+y
print(sumar(3,10))
```

```python
descuentoLambda = lambda desc, precio: ((precio *(1-desc)), (precio * desc))
nuevoPrecio = descuentoLambda(0.4,500)
print("El precio a pagar es", nuevoPrecio[0], ", el descuento que se hizo fue de: ", nuevoPrecio)
print(type(nuevoPrecio))
```
